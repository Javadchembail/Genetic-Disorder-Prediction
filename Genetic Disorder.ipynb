```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
```

```python
x=data.drop(['Genetic Disorder'],axis=1)
y=data['Genetic Disorder']
```

```python
rf_params={'n_estimators':[10,20,30,50]}
```

```python
model_details={'RandomForestClassifier':{'model':RandomForestClassifier(),'params':rf_params},
              'LinearRegression':{'model':LinearRegression(),'params':{}},
              'DecisionTreeClassifier':{'model':DecisionTreeClassifier(),'params':{}}}
```

```python
def get_best(x,y):
    results=[]
    for model_name in model_details:
        model_dict=model_details.get(model_name)
        model=model_dict.get('model')
        params=model_dict.get('params')
        cv=GridSearchCV(model,params,cv=5,return_train_score=True)
        cv.fit(x,y)
        bp=cv.best_params_
        bs=cv.best_score_
        t_res={'model':model_name,'best_params':bp,'best_score':bs}
        results.append(t_res)
    res_df=pd.DataFrame(results)
    best_model=res_df[res_df['best_score']==res_df['best_score'].max()]
    return res_df,best_model
```

```python
df,model=get_best(x,y)
```

```python
df
```

```python
model = RandomForestClassifier()
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=.75,random_state=1)
```

```python
model.fit(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

```python
yact=ytest
ypred=model.predict(xtest)
from sklearn.metrics import confusion_matrix
cf=confusion_matrix(yact,ypred)
```

```python
from mlxtend.plotting import plot_confusion_matrix
plot_confusion_matrix(cf)
```
